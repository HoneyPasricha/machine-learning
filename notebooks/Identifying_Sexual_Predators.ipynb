I will get you started with some code that will parse an XML file:



```python
import xml.etree.ElementTree as ET
counter = 0
def processPredators(filename):
    counter = 0
    tree = ET.parse(filename)
    root = tree.getroot()
    for conversation in root:
        if counter < 20:
            counter +=1
            print('hi', counter)
            for message in conversation:
                author = message.find('author').text
                msg = message.find('text').text
                print('%s: %s' % (author, msg[:80]))
            
# replace the path with yours        
processPredators('/Users/raz/data/pan12-sexual-predator-identification-training-corpus-2012-05-01/pan12-sexual-predator-identification-training-corpus-2012-05-01.xml')
```

This just gets the author ID and the messages out of the first 20 conversations.

# Evaluating
Here is where you need to do some work.


Some people have criticized this dataset as not having a balanced number of predators and non-predators. Rather there are far more non-predators than predators. Let's say that 99% of the people in the dataset are non-predators. We could have a pretty stupid classifier that just said 'not a predator' all the time and be 99% accurate.

For truly evaluating our algorithm we need to use something called the kappa statistic. You will need to read about it starting on page 20 of [chapter 5 of the Guide To Datamining.](http://guidetodatamining.com/assets/guideChapters/DataMining-ch5.pdf)

I would like you to compute the kappa statistic for your algorithm. You can do part of this by hand if you want.
